```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import glob, os
from pandas.plotting import parallel_coordinates
```

```python
data = pd.read_csv('../dataset2/dataset_completo.csv')
```

```python
data.info()
```

```python
data['total_Op']= float(data['total_Op'])
data['total_Opnd'].astype("float64")
data['uniq_Op'].astype("float64")
data['uniq_Opnd'].astype("float64")
```

```python
f,ax = plt.subplots(figsize = (20, 20))
sns.heatmap(data.corr(), annot = True, linewidths = .2, fmt = '.2f')
plt.show()
```

```python
data.head()
```

```python
data.columns
```

```python
g = sns.pairplot(data, hue='defects')
```

```python
f,ax = plt.subplots(figsize = (20, 30))
parallel_coordinates(data, 'defects', colormap=plt.get_cmap("Set2"))
plt.show()
```

```python
newData = data
del newData['McCabe_loc']
```

```python
sns.boxplot(x='defects', y='McCabe_iv(g)', data=data)
```

## Árvore de decisão

```python
from sklearn import tree
```

```python
classes = data['defects']
```

```python
del data['defects']
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(data, classes)
```

```python
nan_rows = data[data.isnull().any(1)]
nan_rows
```

```python
tree.plot_tree(clf.fit(data, classes))
```

```python
import graphviz
import os
```

```python
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'

dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("Software Defective")
```

```python
data.columns
```

```python
graph =dot_data = tree.export_graphviz(clf, out_file=None, 
                      feature_names=np.array(data.columns),  
                      class_names=np.array(['OK', 'Defeito']),  
                      filled=True, rounded=True,  
                      special_characters=True)  
dot_data
graph = graphviz.Source(dot_data)  
graph.render("arvore")
```
